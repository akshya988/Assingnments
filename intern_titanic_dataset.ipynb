```python
# Titanic Dataset - Exploratory Data Analysis (EDA)

# Install libraries (if needed in Colab)
!pip install seaborn matplotlib pandas

# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Preview the data
titanic.head()
```

```python
# Shape of dataset
print("Shape:", titanic.shape)

# Dataset info
print("\nInfo:")
print(titanic.info())

# Missing values
print("\nMissing values:")
print(titanic.isnull().sum())
```

```python
titanic.describe(include='all')
```

```python
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(data=titanic, x='class', hue='survived')
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
sns.histplot(data=titanic, x='age', hue='survived', kde=True, bins=30)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Insights & Observations")
print("- Women had higher survival rate.")
print("- 1st class passengers had better chances of survival.")
print("- Younger children had better survival chances.")
print("- Fare correlates positively with survival.")
```
